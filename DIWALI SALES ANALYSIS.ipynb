EDA

```python
GENDER DISTRIBUTION
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
cp=sns.countplot(x="Gender",data=df)
for bars in cp.containers:
    cp.bar_label(bars)
```

AGE GROUPWISE GENDER DISTRIBUTION

```python
age=sns.countplot(data=df,x='Age Group',hue="Gender")
for bars in age.containers:
   age.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
From the above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
GENDERWISE AMOUNT DISTRIBUTION
```

```python
gender_amount=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(3,5)})
x=sns.barplot(x='Gender',y='Amount',data=gender_amount)
for bars in x.containers:
    x.bar_label(bars)
```

Total orders from top 5 states

```python
top_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
x=sns.barplot(x='State',y='Orders',data=top_orders)
for bars in x.containers:
    x.bar_label(bars)
```

```python
df.columns
```

Top 5 States by sales

```python
top_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
x=sns.barplot(x='State',y='Amount',data=top_sales)
```

OCCUPATION

```python
occupation_order = df['Occupation'].value_counts().index
x=sns.countplot(x='Occupation',data=df,order=occupation_order)
sns.set(rc={'figure.figsize':(35,10)})
for bars in x.containers:
     x.bar_label(bars)
```

```python
occupation_order = df['Occupation'].value_counts().index
occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(40,10)})
x=sns.barplot(x='Occupation',y='Amount',data=occupation,order=occupation_order)
```

PRODUCT CATEGORY

```python
Product_Category= df['Product_Category'].value_counts().index
x=sns.countplot(x='Product_Category',data=df,order=Product_Category)
sns.set(rc={'figure.figsize':(35,10)})
for bars in x.containers:
     x.bar_label(bars)
```

Sales by product category

```python
Product_Category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(40,10)})
x=sns.barplot(x='Product_Category',y='Amount',data=Product_Category)
```
